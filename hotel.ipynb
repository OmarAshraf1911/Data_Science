<h3 style="color:#023e8a; font-size:30px;">Data clean</h3>

```python
mask = data.isnull().any(axis=0)
cols_with_nan = data.columns[mask]
data[cols_with_nan]
```

```python
import math
mostYear=data['arrival_year'].mode()
mostMarket=data['market_segment_type'].mode()
selected_column = 'avg_price_per_room'
col_means = data[selected_column].mean()
col_means_int = math.floor(col_means)
print(mostYear)
print(mostMarket)
print(f"The average is {col_means_int}")
```

```python
data_clean_1=data.fillna({'arrival_year' : 2018.0,'market_segment_type' : 'Online','avg_price_per_room':col_means_int})
```

```python
data_clean_1.isnull().sum() 
```

```python
mask=data_clean_1.isnull().any(axis=1)
num=mask.sum()
num / len(data_clean_1)
```

```python
mask=data_clean_1.isnull().any(axis=1)
data_clean = data_clean_1[~mask]
data_clean.isnull().sum() 
```

```python
# Assuming 'df' is your DataFrame
# To find all duplicate rows based on all columns
duplicates = data_clean.duplicated(keep=False)

# To display only the duplicate rows
duplicate_rows = data_clean[duplicates]

print(duplicate_rows)
```

```python
# To remove all duplicate rows based on all columns
data_clean = data_clean.drop_duplicates(keep=False)
print(data_clean)
```

<h3 style="color:#023e8a; font-size:25px;">Detect and Remove Outliers</h3>

```python
fig = px.histogram(data_clean, x='avg_price_per_room', nbins=30)
fig.show()
```

```python
# to see outliers clearly
fig = px.box(data_clean, y='avg_price_per_room')
fig.show()
```

<h3 style="color:#023e8a; font-size:20px;">Z-score method
</h3>

```python
# find the limits
upper_limit = data_clean['avg_price_per_room'].mean() + 3*data_clean['avg_price_per_room'].std()
lower_limit = data_clean['avg_price_per_room'].mean() - 3*data_clean['avg_price_per_room'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
# find the outliers
data_clean.loc[(data_clean['avg_price_per_room'] > upper_limit) | (data_clean['avg_price_per_room'] < lower_limit)]
```

```python
# trimming - delete the outlier data
new_data_clean = data_clean.loc[(data_clean['avg_price_per_room'] <= upper_limit) & (data_clean['avg_price_per_room'] >= lower_limit)]
print('before removing outliers:', len(data_clean))
print('after removing outliers:',len(new_data_clean))
print('outliers:', len(data_clean)-len(new_data_clean))
```

```python
import plotly.graph_objects as go
fig = go.Figure(data=[go.Box(y=new_data_clean['avg_price_per_room'])])
fig.show()
```

```python
# capping - change the outlier values to upper (or) lower limit values
new_data_clean = data_clean.copy()
new_data_clean.loc[(new_data_clean['avg_price_per_room']>=upper_limit), 'avg_price_per_room'] = upper_limit
new_data_clean.loc[(new_data_clean['avg_price_per_room']<=lower_limit), 'avg_price_per_room'] = lower_limit
```

```python
fig = go.Figure()
fig.add_trace(go.Box(y=new_data_clean['avg_price_per_room']))
fig.update_layout(title='Box Plot of avg_price_per_room')
fig.show()
```
